## Summarization of Calories intake during a day

```python
import matplotlib.pyplot as plt

# Meal categories and their total calories
meals = ['Breakfast', 'Lunch', 'Dinner']
calories = [410, 565, 644]

# Plotting the histogram
plt.bar(meals, calories)
plt.xlabel('Meals')
plt.ylabel('Calories')
plt.title('Calories by Meal')
plt.ylim(0, 700)  # Adjusting y-axis limit for better visualization
plt.show()
```
